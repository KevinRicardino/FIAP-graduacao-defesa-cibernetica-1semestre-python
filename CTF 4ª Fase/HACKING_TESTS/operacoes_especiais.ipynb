```python
class FrutasDict(dict):
    def __delitem__(self, key):
        if key in self:
            super().__delitem__(key)
        else:
            raise keyError("Chave n√£o encontrada.")

frutas = FrutasDict(
    {
        "maca": 2,
        "banana": 3,
        "laranja": 5
    }
)
del frutas["banana"]
frutas["uva"] = 4
print(len(frutas))
```
